### Tracer la surface de la fonction à deux variables f(x,y) = x^2 + y^2

```python
# Définitiuon de la fonction f(x,y) = x^2 + y^2
def f(X, Y):
    return X**2 + Y**2
```

```python
def afficher_surface(f, lower=-10, upper=10, n_points=100):
  X = np.linspace(lower, upper, n_points)
  Y = np.linspace(lower, upper, n_points)

  # Créer le maillage
  X, Y = np.meshgrid(X, Y) # Crée une grille de coordonnées à partir de vecteurs 1D
  Z = f(X, Y) # Calcule les valeurs de la fonction f pour chaque couple de coordonnées

  # Afficher la surface 3D avec Plotly
  fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
  fig.update_layout(width=700, height=700)
  fig.show()
```

## Les Dérivées

**`Le nombre dérivé`** <br>

Le `nombre dérivé`  d'une fonction se calcule en un point par où passe la **tangente** à la courbe de cette fonction. Ce nombre dérivé `est le coefficient directeur de la tangente en ce point`. De plus, `le signe du nombre dérivé nous indique le sens de variation de la fonction`. À savoir, si `le nombre dérivé est négatif` alors la pente de la tangente est décroissante et donc la courbe de `cette fonction décroît` aussi en ce point. Si `le nombre dérivé est positif` alors la pente de la tangente est croissante et donc la courbe de ` cette fonction croît` aussi en ce point. Si `le nombre dérivé est nul` alors la pente de la tangente est parallèle à l'axe des abscisses et donc la courbe de `cette fonction est constante en ce point`. De ce qui précède, cela nous `indique aussi un extremum local (minimum ou maximum)` en ce point. <br><br>
En machine learning (ML), `les fonctions dérivées sont utilisées pour trouver les points d'extremum locaux (minimum ou maximum)` d'une fonction. En effet, `les points d'extremum locaux sont des points où la dérivée de la fonction est nulle`. Cela nous permet de trouver les valeurs des paramètres qui minimisent ou maximisent une `fonction de coût`. <br><br>

**`Les dérivées usuelles à connaître en machine learning :`** <br>
- C' = 0

- x' = 1

- (ax)' = a

- (x²)' = 2x

- [exp(x)]' = exp(x)

- (f o u)' = (f' o u).u' (chain rule)

### Les dérivées partielles, le gradient et le lapacien d'une fonction à deux variables

```python
# Soit la fonction f(x,y) = x^2 * y + exp(x*y)
f = x**2 * y + sp.exp(x*y)

# Calculer la dérivée partielle de f(x,y) par rapport à x
df_dx = sp.diff(f, x)

# Calculer la dérivée partielle de f(x,y) par rapport à y
df_dy = sp.diff(f, y)

# Afficher la dérivée partielle de f(x,y) par rapport à x
print("∂f/∂x = ")
df_dx
```

```python
# Afficher la dérivée partielle de f(x,y) par rapport à y
print("∂f/∂y = ")
df_dy
```

### Le gradient d'une fonction à deux variables

```python
# Calculer le gradient de f(x,y)
# Rappel : Le gradient de f(x,y) est un vecteur ayant pour coordonnées les dérivées partielles de f(x,y) par rapport à x et y dans R².
# sp.Matrix([...]) : SymPy crée une matrice avec une ligne pour chaque valeur.
# sp.Matrix([[...]]) : SymPy crée une matrice sur une même ligne.
gradient_f = sp.Matrix([sp.diff(f, i) for i in (x, y)])

# Afficher le gradient de f(x,y)
print("Gradient de f(x,y) = ")
gradient_f
```

NB: Le gradient est un vecteur qui pointe dans la direction de la plus forte croissance de la fonction. En d'autres termes, le gradient nous indique la direction dans laquelle la fonction croît le plus rapidement. De plus, le gradient est perpendiculaire aux courbes de niveau de la fonction.

### Le laplacien d'une fonction à deux variables

```python
# Calculer le laplacien de f(x,y)
# Rappel : Le laplacien de f(x,y) est la somme des dérivées partielles secondes de f(x,y) par rapport à x et y.
laplacien = sp.diff(f, x, 2) + sp.diff(f, y, 2)

print("Laplacien de f(x,y) = ")
laplacien
```

### Application du gradient et du laplacien en Machine Learning

### Le gradient en Machine Learning
Le gradient est probablement l'opérateur le plus utilisé en machine learning, notamment dans les algorithmes d'optimisation de fonction de coût comme la descente de gradient. En effet, le gradient nous indique la direction dans laquelle la fonction de coût croît le plus rapidement. Ainsi, en suivant le gradient, nous pouvons trouver les valeurs des paramètres qui minimisent la fonction de coût.

```python
# Optimisation d'une fonction de coût simple

# Définir une fonction de coût simple (par exemple, pour une régression logistique)
cost = (1 / (1 + sp.exp(-w1 - w2))) - 0.5

# Calculer le gradient
gradient = sp.Matrix([sp.diff(cost, w1), sp.diff(cost, w2)])

print("Gradient de la fonction de coût = ")
gradient
```

```python
# Deuxième façon d'afficher le résultat
print("Gradient de la fonction de coût = ")
print(gradient)
```

NB: Ce gradient pourrait être utilisé pour mettre à jour les poids w1 et w2 dans un algorithme d'optimisation.

### Le laplacien en Machine Learning


Le laplacien est moins courant mais peut être utilisé dans certaines techniques de régularisation ou dans des architectures de réseaux de neurones spécifiques. Par exemple, le laplacien peut être utilisé pour ajouter du bruit aux images dans le but de les rendre plus robustes aux perturbations. En effet, le laplacien est un opérateur qui permet de détecter les contours dans une image. Ainsi, en ajoutant du bruit laplacien à une image, nous pouvons renforcer les contours de cette image.

```python
# Régularisation Laplacienne pour un modèle simple

# Définir une fonction simple (par exemple, une surface de décision)
f = x**2 + y**2

# Calculer le laplacien
laplacien = sp.diff(f, x, 2) + sp.diff(f, y, 2)

print("Laplacien de la fonction = ")
laplacien
```

NB: Le laplacien pourrait être utilisé comme terme de régularisation pour favoriser des surfaces de décision plus lisses.
